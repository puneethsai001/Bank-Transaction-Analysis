```python
def create_sequences(data, window_size=30):
    sequences = []
    labels = []
    for i in range(len(data) - window_size):
        seq = data[i:i+window_size]
        label = data[i+window_size]
        sequences.append(seq)
        labels.append(label)
    return np.array(sequences), np.array(labels)

# Normalize data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(daily_expenses.values.reshape(-1, 1))

# Create sequences
X, y = create_sequences(scaled_data)
```

```python
# Your sequence creation
X, y = create_sequences(scaled_data)

# Train-test split
train_ratio = 0.8
train_size = int(len(X) * train_ratio)

X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
model = Sequential([LSTM(50, activation='relu', input_shape=(X_train.shape[1], 1)),
                    Dropout(0.2),
                    Dense(1)
                  ])
model.compile(optimizer='adam', loss='mse')
```

```python
history = model.fit(X_train, y_train,epochs=100,batch_size=32,validation_data=(X_test, y_test),verbose=1)
```

```python
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.dates import DateFormatter
from sklearn.metrics import mean_squared_error


# Assuming daily_expenses, scaled_data, X_test, y_test, model, and scaler are already defined and trained from previous cells.

# Make predictions on the test data
predictions = model.predict(X_test)
predictions_rescaled = scaler.inverse_transform(predictions.reshape(-1, 1))
y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1, 1))

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test_rescaled, predictions_rescaled))
print(f"RMSE on Test Data: Dhs {rmse:.2f}")


def get_forecast(transaction_df, scaler, model):
    # Your preprocessing pipeline here
    daily_expenses = transaction_df[transaction_df['Amount'] < 0]['Amount'].resample('D').sum().abs()
    scaled_data = scaler.transform(daily_expenses.values.reshape(-1, 1))
    last_window = scaled_data[-30:]
    # Assuming predict_future is defined elsewhere
    return predict_future(model, last_window, scaler=scaler)


def predict_future(model, last_sequence, steps=30, scaler=None):
    future_predictions = []
    current_sequence = last_sequence.copy()

    for _ in range(steps):
        # Reshape current_sequence for model prediction
        next_pred = model.predict(current_sequence.reshape(1, current_sequence.shape[0], 1), verbose=0)
        future_predictions.append(next_pred[0,0])
        # Update current_sequence for the next step
        current_sequence = np.roll(current_sequence, -1)
        current_sequence[-1] = next_pred[0,0]

    if scaler is not None:
         # Ensure the input to inverse_transform is 2D
        return scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))
    else:
        return np.array(future_predictions).reshape(-1, 1)


def visualize_forecast(transaction_df, rmse, scaler, model, forecast_days=30): # Added rmse, scaler, model as arguments
    """
    Visualizes historical expenses and LSTM forecasts with interactive elements.
    """
    # Get forecasts - Pass scaler and model
    forecasts = get_forecast(transaction_df, scaler, model)

    # Prepare data
    daily_expenses = transaction_df[transaction_df['Amount'] < 0]['Amount'].resample('D').sum().abs()
    last_60_days = daily_expenses[-60:]
    # Ensure last_60_days is not empty
    if last_60_days.empty:
        print("Warning: Not enough historical data to display last 60 days.")
        last_60_days = daily_expenses
        if last_60_days.empty:
             print("Error: No daily expenses data available.")
             return None # Or handle appropriately

    # Ensure daily_expenses index is a DatetimeIndex before calculation
    if not isinstance(daily_expenses.index, pd.DatetimeIndex):
         daily_expenses.index = pd.to_datetime(daily_expenses.index)

    # Ensure daily_expenses is not empty before calculating forecast_dates
    if daily_expenses.empty:
        print("Error: No daily expenses data available to determine forecast start date.")
        return None # Or handle appropriately


    # Calculate forecast_dates safely
    start_date = daily_expenses.index[-1]
    forecast_dates = pd.date_range(start_date, periods=forecast_days + 1, freq='D')[1:]


    # Create figure
    plt.figure(figsize=(14, 7))

    # Style
    sns.set_style("whitegrid")
    plt.rcParams['font.size'] = 12

    # Plot historical data
    plt.plot(last_60_days.index,
             last_60_days.values,
             label='Historical Expenses',
             color='#1f77b4',
             linewidth=2,
             marker='o',
             markersize=5)

    # Plot forecast
    plt.plot(forecast_dates,
             forecasts,
             label='Forecasted Expenses',
             color='#ff7f0e',
             linewidth=2,
             linestyle='--',
             marker='x',
             markersize=6)

    # Confidence interval (example: ±15%)
    plt.fill_between(forecast_dates,
                    forecasts.flatten() * 0.85,
                    forecasts.flatten() * 1.15,
                    color='#ff7f0e',
                    alpha=0.2,
                    label='Confidence Range (±15%)')

    # Formatting
    plt.title(f'{forecast_days}-Day Expense Forecast\nRMSE: Dhs{rmse:.2f}', pad=20, fontsize=16)
    plt.xlabel('Date', labelpad=10)
    plt.ylabel('Daily Expenses', labelpad=10)

    # Date formatting
    date_form = DateFormatter("%b %d")
    plt.gca().xaxis.set_major_formatter(date_form)
    plt.xticks(rotation=45)

    # Annotations - Ensure forecasts is not empty before calculating max
    if forecasts.size > 0:
        max_forecast_day = forecast_dates[np.argmax(forecasts)]
        plt.annotate(f'Peak: Dhs{forecasts.max():.2f} on {max_forecast_day.strftime("%b %d")}',
                    xy=(max_forecast_day, forecasts.max()),
                    xytext=(10, 10),
                    textcoords='offset points',
                    arrowprops=dict(arrowstyle="->"))

    # Legend and grid
    plt.legend(loc='upper left')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()

    # Interactive elements (if using Jupyter)
    plt.gcf().canvas.toolbar_visible = True
    plt.gcf().canvas.header_visible = False

    return plt

# Usage
chart = visualize_forecast(transaction_df, rmse, scaler, model, forecast_days=30) # Pass rmse, scaler, and model
if chart:
    chart.show()
```

```python
def detect_monthly_anomaly(transaction_row):
    credit = transaction_row.get('Credit', 0)
    debit = transaction_row.get('Debit', 0)
    balance = transaction_row.get('Balance', 0)
    date_str = transaction_row.get('Date')  # e.g., '2025-07-15'
    amount = credit - debit

    try:
        input_month = pd.to_datetime(date_str).to_period('M')
        month_key = input_month.strftime('%Y-%m')
        model = monthly_models.get(month_key)

        # If month not found (i.e., future), use most recent available model
        if not model:
            latest_month_key = sorted(monthly_models.keys())[-1]
            print(f"[Info] Using latest available model from: {latest_month_key}")
            model = monthly_models[latest_month_key]

        features_df = pd.DataFrame([[amount, balance]], columns=['Amount', 'Balance'])
        prediction = model.predict(features_df)
        return 'Anomaly' if prediction[0] == -1 else 'Normal'
    except Exception as e:
        return f"Error: {e}"
```

```python
# Step 1: Identify off-track months
off_track_months = monthly_summary[monthly_summary['On_Track'] == False]['Month'].tolist()

# Step 2: Filter transactions in off-track months AND are expenses only (Amount < 0)
expense_data = new_df[
    (new_df['Month'].astype(str).isin(off_track_months)) &
    (new_df['Amount'] < 0)
]

# Step 3: Exclude known income categories just in case they're mislabeled
excluded_income_categories = ['Salary', 'Utilities', 'EMI', 'Insurance', 'Healthcare']
expense_data = expense_data[~expense_data['Category'].isin(excluded_income_categories)]

# Step 4: Group by Month and Category, get total spending
category_wise_expense = expense_data.groupby(['Month', 'Category'])['Amount'].sum().abs().reset_index()

# Step 5: Sort to show top spending categories
category_wise_expense = category_wise_expense.sort_values(['Month', 'Amount'], ascending=[True, False])

# Step 6: Display result
print(category_wise_expense)
```
